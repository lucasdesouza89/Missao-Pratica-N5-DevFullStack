```python
#Importar bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
!pip install faker
!pip install num2words
from num2words import num2words
from faker import Faker
from pylab import *
```

```python
#Gerar função para criação de nomes e notas falsas
fk = Faker('pt-br')
def gerar_nomes(quantos):
  nomes = []
  for _ in range(0, quantos):
    nomes.append(fk.name())
    nomes.append(fk.random_int(1,10))
  return nomes
```

```python
#Salvar dados gerados em arquivo
with open('dados.txt', 'w', newline='') as file:
  for i in gerar_nomes(50):
    file.write(str([i]) + '\n')
```

```python
#Abrir dados salvos
with open('dados.txt', 'r') as arquivo:
  for i, linha in enumerate(arquivo):
    if i % 2 == 0:
      print("Nome", linha, end='')
    else:
      print("Nota", linha, end='')
```

```python
#Gerar histograma
with open('dados.txt', 'r') as arquivo:
  notas = []
  for i, linha in enumerate(arquivo):
    if i % 2 == 1:
      notas.append(linha)
      notas.sort()

hist(notas)
plt.title('Histograma das Pontuações')
plt.ylabel('Probabilidade')
plt.xlabel('Pontuações')
plt.grid()
plt.show()
```

```python
#Gerar núvem de palavras
palavras = ' '.join(sorted(notas))
a = palavras.replace('10','0')
palavras1 = []
for i in a:
  if '1' in i:
     palavras1.append(num2words(1,lang='pt-br'))
  if "2" in i:
    palavras1.append(num2words(2,lang='pt-br'))
  if "3" in i:
    palavras1.append(num2words(3,lang='pt-br'))
  if "4" in i:
    palavras1.append(num2words(4,lang='pt-br'))
  if "5" in i:
    palavras1.append(num2words(5,lang='pt-br'))
  if "6" in i:
    palavras1.append(num2words(6,lang='pt-br'))
  if "7" in i:
    palavras1.append(num2words(7,lang='pt-br'))
  if "8" in i:
    palavras1.append(num2words(8,lang='pt-br'))
  if "9" in i:
    palavras1.append(num2words(9,lang='pt-br'))
  if "0" in i:
    palavras1.append(num2words(10,lang='pt-br'))

b = ' '.join(palavras1)

nuvem_de_palavras = WordCloud().generate(b)
plt.figure(figsize=(15,8))
plt.axis("off")
plt.imshow(nuvem_de_palavras, interpolation='bilinear')
plt.show()
nuvem_de_palavras.to_file("Nuvem de palavras.png")

```
